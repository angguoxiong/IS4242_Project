```python
import pandas as pd
import numpy as np
import json
```

# Dataset

```python
df1 = pd.read_csv('movies_metadata.csv')
df2 = pd.read_csv('ratings.csv')
```

```python
df1.head(5)
```

```python
df2.head(5)
```

#### Merge User Rating with Movies 


```python
df1 = df1[~df1['id'].isin(['2014-01-01', '2012-09-29', '1997-08-20'])]
df1['id'] = df1['id'].astype(int)
df_merged = df2.merge(df1, left_on = 'movieId', right_on = 'id', how = 'left')
```

```python
df_merged.head(5)
```

```python
# create boolean mask of NaN values
nan_mask = df_merged.isna()

# count number of True values in each column
nan_count = nan_mask.sum(axis=0)

# print number of NaN values in each column
print(nan_count/len(df_merged))
```

```python
df_merged.columns
```

```python
df_final = df_merged.drop(columns = ['timestamp', 'budget', 'homepage', 'id',
                                     'imdb_id', 'overview', 'poster_path', 'tagline', 'movieId','original_title', 'title',
                                    'belongs_to_collection', 'video','title'])
```

```python
df_final.head(5)
```

# Data Cleaning

#### Extract genres from dictionary

```python
df_final['genres_list'] = df_final['genres'].apply(lambda row: [d['name'] for d in json.loads(row.replace("'", "\""))] if isinstance(row, str) else [])
```

#### Extract num_years since movie released

```python
from datetime import datetime

df_final['years_since_released'] = df_final['release_date'].apply(lambda x: x if pd.isna(x) else 2023-datetime.strptime(str(x), "%Y-%m-%d").year)
```

```python
df_final.drop(columns = ['genres', 'release_date'], inplace = True)
df_final.head(5)
```

#### Replace True False with Binary Value

```python
df_final['adult'] = df_final['adult'].replace({"True": 1, "False": 0})
df_final['video'] = df_final['video'].replace({"True": 1, "False": 0})
```
